```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
import seaborn as sns
import matplotlib.pyplot as plt

# === Load datasets ===
peptide_df = pd.read_csv("peptide_features.csv")
protein_df = pd.read_csv("protein_features.csv")

# === Merge peptide and protein features ===
merged_df = pd.merge(peptide_df, protein_df, on="protein_id", suffixes=("_pep", "_prot"))

PRESENCE_LABELS = {
    1: "Peptide predicted, neither peptide protein not seen",
    2: "Peptide predicted, protein seen but peptide not seen",
    3: "Peptide & protein predicted and Seen",
    4: "Peptide seen under other protein Id",
    5: "Peptide seen, but not predicted",
    6: "Peptide seen only in other experiment",
}

# ------------------------------------------------------------------
# CLEAN BINARIZATION
# 0 = "not observed"  categories 1, 2
# 1 = "observed"      categories 3, 4, 5, 6
# ------------------------------------------------------------------
def binarize_presence(s):
    return s.replace({1:0, 2:0, 3:1, 4:1, 5:1, 6:1})

merged_df["presence_SILAC_bin"] = binarize_presence(merged_df["presence_SILAC"])
merged_df["presence_Label_free_bin"] = binarize_presence(merged_df["presence_Label_free"])

# Ensure int dtype
merged_df["presence_SILAC_bin"] = merged_df["presence_SILAC_bin"].astype(int)
merged_df["presence_Label_free_bin"] = merged_df["presence_Label_free_bin"].astype(int)

print("Unique SILAC bin labels:", merged_df["presence_SILAC_bin"].unique())
print("Unique Label-free bin labels:", merged_df["presence_Label_free_bin"].unique())

# === Define labels ===
y_silac = merged_df["presence_SILAC_bin"]
y_label_free = merged_df["presence_Label_free_bin"]

# === Select feature columns ===
exclude_cols = [
    "protein_id", "peptide", "protein_sequence_pep", "protein_sequence_prot",
    "presence_SILAC", "presence_Label_free", "presence_SILAC_bin", "presence_Label_free_bin",
    "observed_frequency_SILAC", "observed_intensity_sum_SILAC",
    "observed_frequency_Label_free", "observed_intensity_sum_Label_free"
]

X = merged_df.drop(columns=[c for c in exclude_cols if c in merged_df.columns])
X = X.fillna(0)

# === Split train/val/test ===
X_train, X_temp, y1_train, y1_temp, y2_train, y2_temp = train_test_split(
    X, y_silac, y_label_free, test_size=0.3, random_state=42, stratify=y_silac
)
X_val, X_test, y1_val, y1_test, y2_val, y2_test = train_test_split(
    X_temp, y1_temp, y2_temp, test_size=0.5, random_state=42
)

# === Scale ===
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# === Models ===
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=42),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss'
    ),
    "SVM": SVC(kernel='rbf', C=2, probability=True),
    "MLP": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
}

def evaluate_models(X_train, X_val, y_train, y_val, label_name):
    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        preds = model.predict(X_val)
        acc = accuracy_score(y_val, preds)
        # explicit binary F1 with pos_label=1
        f1 = f1_score(y_val, preds, average='binary', pos_label=1, zero_division=0)
        results.append((name, acc, f1))

        print(f"\n=== {label_name} - {name} ===")
        print(f"Accuracy: {acc:.3f}, F1-score: {f1:.3f}")
        print("Classification report:")
        print(classification_report(y_val, preds, zero_division=0))
        cm = confusion_matrix(y_val, preds, labels=[0, 1])
        plt.figure(figsize=(4,3))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0,1], yticklabels=[0,1])
        plt.title(f"{label_name} Confusion Matrix - {name}")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.tight_layout()
        plt.show()

    df_results = pd.DataFrame(results, columns=["Model", "Accuracy", "F1"])
    print(f"\nSummary for {label_name}:\n", df_results.sort_values("F1", ascending=False))
    return df_results


# === Evaluate both labels ===
print("\n########## Evaluating presence_SILAC (binary) ##########")
results_silac = evaluate_models(X_train_scaled, X_val_scaled, y1_train, y1_val, "presence_SILAC_bin")

print("\n########## Evaluating presence_Label_free (binary) ##########")
results_label_free = evaluate_models(X_train_scaled, X_val_scaled, y2_train, y2_val, "presence_Label_free_bin")

# === Example predictions ===
best_model = RandomForestClassifier(n_estimators=300, random_state=42)
best_model.fit(X_train_scaled, y1_train)
sample_preds = best_model.predict(X_test_scaled[:10])

print("\nExample binary predictions (presence_SILAC):")
print("True:", list(y1_test[:10].values))
print("Pred:", list(sample_preds))
```
